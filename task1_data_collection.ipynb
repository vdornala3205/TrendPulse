# Task 1: Make the API Calls

```python
import requests
from bs4 import BeautifulSoup
```

```python


# Step 1 — Get the list of top story IDs

url = 'https://hacker-news.firebaseio.com/v0/topstories.json'
response = requests.get(url)
top_story_ids = response.json()
data = response.json()

print(data)
```

```python
# Fetch 500 IDs from the returned array

first_500_story_ids = top_story_ids[:500]
display(first_500_story_ids)
```

```python
# Step 2 — Get each story details; add headers

story_details = []
base_item_url = 'https://hacker-news.firebaseio.com/v0/item/{}.json'
headers = {"User-Agent": "TrendPulse/1.0"}

for story_id in first_500_story_ids:
    item_url = base_item_url.format(story_id)
  # adding a try-except block to handle failed requests
    try:
        response = requests.get(item_url, headers=headers)
        response.raise_for_status() # Raise an exception for bad status codes (4xx or 5xx)
        story_details.append(response.json())
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch story {story_id}: {e}")

# Verify the first 5 story details
display(story_details[:5])
```

```python
# Assign a category to each story by checking whether its title contains any of these keywords:

import pandas as pd

categories = {
    'technology': ['AI', 'software', 'tech', 'code', 'computer', 'data', 'cloud', 'API', 'GPU', 'LLM'],
    'worldnews': ['war', 'government', 'country', 'president', 'election', 'climate', 'attack', 'global'],
    'sports': ['NFL', 'NBA', 'FIFA', 'sport', 'game', 'team', 'player', 'league', 'championship'],
    'science': ['research', 'study', 'space', 'physics', 'biology', 'discovery', 'NASA', 'genome'],
    'entertainment': ['movie', 'film', 'music', 'Netflix', 'game', 'book', 'show', 'award', 'streaming']
}

# Function to assign category
def assign_category(title):
    if title is None: # If a title is missing
        return 'Other'
    title_lower = title.lower()
    for category, keywords in categories.items():
        for keyword in keywords:
            if keyword.lower() in title_lower:
                return category
    return 'Other'

# Assign categories to each story
for story in story_details:
    story['category'] = assign_category(story.get('title'))

# Display the first 5 stories with their new category
display(pd.DataFrame(story_details[:5]))
```

```python
# Group stories by category and then iterate through those groups with 2 seconds time delay.

import time
from collections import defaultdict

categorized_stories = defaultdict(list)
for story in story_details:
    categorized_stories[story['category']].append(story)

for category, stories in categorized_stories.items():
    print(f"\n--- Category: {category.upper()} ---")
    for i, story in enumerate(stories[:3]): # Displaying first 3 stories from each category
        print(f"  Title: {story.get('title')}")
        if story.get('url'):
            print(f"  URL: {story.get('url')}")
        if i < 2 and len(stories) > 1: # Print a separator for readability
            print("  ------")
    time.sleep(2) # Wait 2 seconds before processing the next category
```

# Task 2: Extract the Fields

```python
#  2. Extract the Fields

import datetime

processed_stories = []

for category, stories in categorized_stories.items():
    stories_in_category = 0
    for story in stories:
        if stories_in_category >= 25:
            break

        # Extract relevant fields
        processed_story = {
            'post_id': story.get('id'),
            'title': story.get('title'),
            'category': story.get('category'),
            'score': story.get('score'),
            'num_comments': story.get('descendants'),
            'author': story.get('by'),
            'collected_at': datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") # Current date and time
        }
        processed_stories.append(processed_story)
        stories_in_category += 1

# Display the first 15 collected stories to verify
display(pd.DataFrame(processed_stories).head(15))
```

# Task 3: Save to JSON File

```python
# 3. Saving to a JSON file

import os
import datetime

# Create the 'data' folder if it doesn't exist already
data_dir = 'data/'
os.makedirs(data_dir, exist_ok=True)

# Generate the filename with the current date
current_date = datetime.datetime.now().strftime("%Y%m%d")
filename = f"trends_{current_date}.json"
file_path = os.path.join(data_dir, filename)

# Convert the list of dictionaries to a pandas DataFrame and save to JSON
df_processed_stories = pd.DataFrame(processed_stories)
df_processed_stories.to_json(file_path, orient='records', indent=4)

print(f"All collected stories saved to: {file_path}")
print(f"Total stories collected: {len(processed_stories)}")
```

```python
# Verify the contens of saved JSON file

import pandas as pd

# Read the JSON file into a pandas DataFrame
df_read = pd.read_json(file_path)

# Display the first few rows of the DataFrame
display(df_read.head())
```

The "Processed Stories" list now contains up to 25 stories for each category with the specified fields and a "collected_at" timestamp.
