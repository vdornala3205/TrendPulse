# 1. Load the JSON File

```python
import pandas as pd

# Pick the JSON file from 'data/' folder that's saved during Task 1.
try:
    df = pd.read_json('/content/drive/MyDrive/Colab Notebooks/trends_20260414.json')
    print(f"\nSuccessfully loaded {df.shape[0]} rows from trends_20260414.json")

except FileNotFoundError:
    print("Error: '/content/drive/MyDrive/Colab Notebooks/trends_20260414.json' not found. Please ensure the file exists in the specified path.")
except Exception as e:
    print(f"An error occurred while loading the JSON file: {e}")
```

# 2. Clean the Data

### Remove Duplicates



```python
print(f"Initial number of rows: {df.shape[0]}")
df.drop_duplicates(subset='post_id', inplace=True)
print(f"Number of rows after removing duplicate post_ids: {df.shape[0]}")
display(df.head())
```

### Handle Missing Values



```python
print(f"Number of rows before dropping rows with missing critical values: {df.shape[0]}")
df.dropna(subset=['post_id', 'title', 'score'], inplace=True)
print(f"Number of rows after dropping rows with missing critical values: {df.shape[0]}")
display(df.head())
```

### Convert Data Types

```python
print("DataFrame info before type conversion:")
df.info()

# Convert 'score' to integer, coercing errors to NaN
df['score'] = pd.to_numeric(df['score'], errors='coerce')
# Convert 'num_comments' to integer, coercing errors to NaN
df['num_comments'] = pd.to_numeric(df['num_comments'], errors='coerce')

# Drop rows where 'score' or 'num_comments' became NaN due to coercion errors
df.dropna(subset=['score', 'num_comments'], inplace=True)

# Convert to integer type after handling NaNs
df['score'] = df['score'].astype(int)
df['num_comments'] = df['num_comments'].astype(int)

print("\nDataFrame info after type conversion:")
df.info()
display(df.head())
```

### Filter Low Quality Stories

```python
print(f"Number of rows before filtering by score: {df.shape[0]}")
df = df[df['score'] >= 5]
print(f"Number of rows after filtering for score >= 5: {df.shape[0]}")
display(df.head())
```

### Clean Whitespace in Title

```python
print("Sample titles before stripping whitespace:")
print(df['title'].head().tolist())

df['title'] = df['title'].str.strip()

print("\nSample titles after stripping whitespace:")
print(df['title'].head().tolist())
display(df.head())
```

### Gist of Cleaned Data

```python
print(f"Final number of rows: {df.shape[0]}")
df.info()
display(df.head())
```

# 3. Save as CSV

```python
import os

# Define the directory for saving the clean data
output_dir = 'data/'

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Define the output file path
output_file_path = os.path.join(output_dir, 'trends_clean.csv')

# Save the cleaned DataFrame to CSV
df.to_csv(output_file_path, index=False)

print(f"Successfully saved {df.shape[0]} rows to {output_file_path}")
```
